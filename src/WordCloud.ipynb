### Draw WordCloud

```python
import warnings
warnings.filterwarnings("ignore",category=DeprecationWarning)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import PIL
```

```python
with open("../corpus/pos.txt", "r", encoding='utf-8') as f:
    pos_text = f.read()   #Open the positive corpus
```

```python
import jieba
STOPWORDS = [
    u'我', u'你', u'的', u'了', u'是', u'有', u'觉得',
    u'啊', u'吧', u'吗', u'就是', u'还是', u'感觉', u'看', u'说', u'不',
    u'很', u'真', u'真的', u'太', u'挺', u'给', u'做', u'非常', u'这', u'就', u'都',
    u'也', u'和', u'但', u'在', u'对', u'能', u'还'
]

wc_zh_pos = WordCloud(font_path='simfang.ttf', stopwords=STOPWORDS, width=1600, height=800)
wc_zh_pos.generate(" ".join(jieba.lcut(pos_text ,cut_all=True, HMM=True)))

plt.imshow(wc_zh_pos)
plt.axis("off")
plt.show()
```

```python
wc_zh_pos.to_file("tb_pos_cloud.png")
```

```python
with open("../corpus/neg.txt", "r", encoding='utf-8') as f:  #打开文本
    neg_text = f.read()   #读取文本
```

```python
wc_zh_neg = WordCloud(font_path='simfang.ttf', stopwords=STOPWORDS, width=1600, height=800)
wc_zh_neg.generate(" ".join(jieba.lcut(neg_text ,HMM=True,cut_all=False)))

plt.imshow(wc_zh_neg)
plt.axis("off")
plt.show()
```

```python
wc_zh_neg.to_file("../pic/tb_fackneg_cloud.png")
```

```python
wc_zh_neg = WordCloud(font_path='simfang.ttf', 
                      stopwords=STOPWORDS+['好','好看','好好','不错','期待','不是', u'动画', u'三体', u'捏'], 
                      width=1600, height=800)
wc_zh_neg.generate(" ".join(jieba.lcut(neg_text , HMM=True ,cut_all=False)))

plt.imshow(wc_zh_neg)
plt.axis("off")
plt.show()
```

```python
wc_zh_neg.to_file("../pic/tb_neg_cloud.png")
```
