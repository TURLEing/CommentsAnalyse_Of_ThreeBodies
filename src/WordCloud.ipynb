### Draw WordCloud

```python
import warnings
warnings.filterwarnings("ignore",category=DeprecationWarning)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import PIL
```

```python
with open("../corpus/pos.txt", "r", encoding='utf-8') as f:
    pos_text = f.read()   #Open the positive corpus
```

```python
import jieba
STOPWORDS = [
    u'', u'', u'', u'', u'', u'', u'',
    u'', u'', u'', u'', u'', u'', u'', u'', u'',
    u'', u'', u'', u'', u'', u'', u'', u'', u'', u'', u'',
    u'', u'', u'', u'', u'', u'', u''
]

wc_zh_pos = WordCloud(font_path='simfang.ttf', stopwords=STOPWORDS, width=1600, height=800)
wc_zh_pos.generate(" ".join(jieba.lcut(pos_text ,cut_all=True, HMM=True)))

plt.imshow(wc_zh_pos)
plt.axis("off")
plt.show()
```

```python
wc_zh_pos.to_file("tb_pos_cloud.png")
```

```python
with open("../corpus/neg.txt", "r", encoding='utf-8') as f:  #
    neg_text = f.read()   #
```

```python
wc_zh_neg = WordCloud(font_path='simfang.ttf', stopwords=STOPWORDS, width=1600, height=800)
wc_zh_neg.generate(" ".join(jieba.lcut(neg_text ,HMM=True,cut_all=False)))

plt.imshow(wc_zh_neg)
plt.axis("off")
plt.show()
```

```python
wc_zh_neg.to_file("../pic/tb_fackneg_cloud.png")
```
